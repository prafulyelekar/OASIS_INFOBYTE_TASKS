# Task-2 :Unemployment Data Analysis

Unemployment is measured by the unemployment rate which is the number of people who are unemployed as a percentage of the total labour force. We have seen a sharp increase in the unemployment rate during Covid-19, so we are analyzing the unemployment rate.
Unemployment Data Analysis
Unemployment data analysis involves the process of collecting, cleaning, analyzing, and interpreting data related to unemployment rates and trends. The analysis of this data can provide valuable insights into the state of the economy and the job market, and help policymakers and businesses make informed decisions.

Steps involved in unemployment data analysis:

STEP-1 : Data collection: Unemployment data can be collected from various sources such as government agencies, private organizations, and academic research studies. The data may include various metrics such as the number of unemployed individuals, the unemployment rate, labor force participation rate, etc.

STEP-2 : Data cleaning: Raw data may contain errors, missing values, and inconsistencies that need to be addressed before analysis. Data cleaning involves identifying and correcting errors, filling in missing values, and removing any outliers that could skew the results.

STEP-3 : Data analysis: Once the data is cleaned, it can be analyzed using various statistical techniques such as regression analysis, time series analysis, and descriptive statistics. This analysis can help identify patterns, trends, and relationships in the data.

STEP-4 : Data visualization: Data visualization techniques such as charts, graphs, and maps can help present the analysis results in an easy-to-understand format.

STEP-5 : Interpretation and reporting: The final step is to interpret the results and report the findings to the relevant stakeholders. This may include policymakers, businesses, and the general public.

# Look at the estimated number of employees according to different regions of India

```python
# Look at the estimated number of employees according to different regions of India
print("This is the estimated number of employees according to different regions of India")
print("This plot is showing the relation between estimated number of employees according to different regions of India Vs Count")
data2.columns = ["States","Date","Frequency","Estimated Unemployment Rate",
                "Estimated Employed","Estimated Labour Participation Rate",
                "Region","longitude","latitude"]

plt.title("Indian Employment")
sns.histplot(x="Estimated Employed",hue="Region",data=data2)
```

```python
# See the unemployment rate according to different regions of India

print("This is the unemployment rate according to different regions of India")
print("This plot is showing the relation between unemployment rate according to different regions of India Vs Count")
plt.figure(figsize=(6,4))
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Unemployment Rate",hue="Region",data=data2)
```

```python
sns.countplot(x='Region',data=data2)
plt.ylabel('Estimated Unemployment Rate (%)')
```

```python
freq = data2['Region'].value_counts()
print(freq)
freq.plot(kind='pie',startangle = 140)
plt.legend()
plt.show()
```

```python
# Now Create a dashboard to analyze the unemployment rate of each Indian state by region --> by using subburst plot

print("Creating a dashboard to analyze the unemployment rate of each Indian state by region")
unemployment = data2[["States","Region","Estimated Unemployment Rate"]]
# figure = px.sunburst()
figure = px.sunburst(unemployment,path=["Region","States"],
                    values="Estimated Unemployment Rate",
                    width=700,height=700,color_continuous_scale="RdY1Gn",
                    title="Unemployment Rate in India (Dashboard to analyze the unemployment rate)")

figure.show()
```
